# Attributes
Attributes can be considered like instance variables in other languages. In python, invoking dir against an object lists all attributes and even lists methods as attributes

```python
class Person(object):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

p = Person("Alice", 30)
print(p.greet())
dir(p)
```

# Properties
Properties in python is a simplified version of "descriptors"

The idea is to protet direct access to an attrobute of a class.
- Setting a value invokes a method
- Getting value invokes a separate method

```python
class BadAgeError(Exception):
    pass

class Person(object):
    def __init__(self, name, age):
        self.name = name
        self._age = age  # Use a protected attribute for age
    
    @property
    def age(self):
        print("Accessing age property")
        return self._age
    
    @age.setter
    def age(self, value):
        if value < 0:
            raise BadAgeError("Age cannot be negative")
        print("Setting age property")
        self._age = value

    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

p = Person("Alice", 30)
print(p.greet())
p.age = 31  # This will call the setter
p.age  # This will call the getter
p.age = -10
```
