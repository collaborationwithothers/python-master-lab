```python
from enum import Enum

class DayOfWeek(Enum):
    SUNDAY = (0, "SUN", True)
    MONDAY = (1, "MON", False)
    TUESDAY = (2, "TUE", False)
    WEDNESDAY = (3, "WED", False)
    THURSDAY = (4, "THU", False)
    FRIDAY = (5, "FRI", False)
    SATURDAY = (6, "SAT", True)

    def __new__(cls, dow_index, abbreviation, is_weekend):
        obj = object.__new__(cls)
        obj._value_ = dow_index
        obj.abbreviation = abbreviation
        obj.is_weekend = is_weekend
        return obj

my_day = DayOfWeek.FRIDAY
print(f"My day is {my_day.name}, abbreviation: {my_day.abbreviation}, is weekend: {my_day.is_weekend}, index: {my_day.value}")
```
