```python
class Duck(object):
    def __init__(self, name):
        self.name = name

    def quack(self):
        return f"{self.name} says Quack!"

class Person(object):
    def __init__(self, name, occuptation):
        self.name = name
        self.occuptation = occuptation

    def quack(self):
        return f"{self.name} says Quack!"
ducks = [Duck("Donald"), Person("John", "Engineer")]

for duck in ducks:
    print(duck.quack())
```

```python
class File(object):
    def __init__(self, name):
        self.name = name

    def read(self):
        print(f"Reading from {self.name}")
    
    def close(self):
        print(f"Closing {self.name}")

class Database(object):
    def __init__(self, name):
        self.name = name

    def read(self):
        print(f"Reading from {self.name}")
    
    def close(self):
        print(f"Closing {self.name}")

class WebService(object):
    def __init__(self, name):
        self.name = name

    def read(self):
        print(f"Reading from {self.name}")
    
    def close(self):
        print(f"Closing {self.name}")

def process_resource(resource):
    resource.read()
    resource.close()

text_file = File("example.txt")
database = Database("example.db")
service = WebService("example.com")

process_resource(text_file)
process_resource(database)
process_resource(service)
```
