```python
def somefunc():
    return 1
    return 2
    return 3

somefunc()
```

```python
def somefunc1():
    yield 1
    yield 2
    yield 3
somefunc1()

for i in somefunc1():
    print(i)
```

```python
def somefunc2():
    print("Before first yield")
    yield 1
    print("Before second yield")
    yield 2
    print("Before third yield")
    yield 3
g = somefunc2()
print(next(g))  # Output: Before first yield, 1
print(next(g))  # Output: Before second yield, 2
print(next(g))  # Output: Before third yield, 3
```

```python
def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

print("Fibonacci sequence:")
for i, value in enumerate(fib()):
    if i >= 10:  # Limit to first 10 Fibonacci numbers
        break
    print(value)
```

```python
def char_range(start='a', end='z'):
    for c in range(ord(start), ord(end)+1):
        yield chr(c)

print("Character range from 'a' to 'z':")
for char in char_range():
    print(char)
```
