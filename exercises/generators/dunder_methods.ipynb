```python
class Employee(object):
    def __init__(self,name,employee_id, salary):
        self.name = name
        self.employee_id = employee_id
        self.salary = salary
    
    def __str__(self):
        return f"Employee Name: {self.name}, Employee ID: {self.employee_id}"
    
    def __repr__(self):
        return f"Employee(name=\"{self.name}\", employee_id=\"{self.employee_id}\", salary=\"{self.salary}\")"
    
    def __eq__(self, other):
        if isinstance(other, Employee):
            return self.employee_id == other.employee_id
        return False
    
    def __lt__(self, other):
        if isinstance(other, Employee):
            return self.salary < other.salary
        return NotImplemented

class Manager(Employee):
    def __init__(self, name, employee_id, salary, team_size):
        super().__init__(name, employee_id, salary)
        self.team_size = team_size
    
    def __str__(self):
        base_str = super().__str__()
        return f"{base_str} - Managing a team of {self.team_size} employees"
    
emp = Employee("John Doe", 12345, 50000)
mgr = Manager("Jane Smith", 54321, 80000, 10)
staff = [emp, mgr]
for emp in staff:
    print(emp)  # Calls __str__

print(emp)  # Calls __str__
print(repr(emp))  # Calls __repr__

emp1 = Employee("Alice", 11111, 60000)
emp2 = Employee("Bob", 11111, 70000)
print(f"Are emp1 and emp2 equal? {emp1 == emp2}")  # True, because employee_id is the same
print(f"Is emp1's salary less than emp2's salary? {emp1 < emp2}")  # True, because salary of emp1 is less than emp2
```

```python
class Book(object):
    def __init__(self, title, author, isbn, pages):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.pages = pages
    
    def __str__(self):
        return f"Book Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}, Pages: {self.pages}"
    
    def __repr__(self):
        return f"Book(title=\"{self.title}\", author=\"{self.author}\", isbn=\"{self.isbn}\", pages={self.pages})"
    
    def __eq__(self, other):
        if isinstance(other, Book):
            return self.isbn == other.isbn
        return False

class EBook(Book):
    def __init__(self, title, author, isbn, pages, file_size):
        super().__init__(title, author, isbn, pages)
        self.file_size = file_size
    
    def __str__(self):
        base_str = super().__str__()
        return f"{base_str} - File Size: {self.file_size} MB"

sample_books = [
    Book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565", 180),
    EBook("1984", "George Orwell", "9780451524935", 328, 1.5),
    Book("To Kill a Mockingbird", "Harper Lee", "9780061120084", 281),
    EBook("Brave New World", "Aldous Huxley", "9780060850524", 268, 2.0)
]

for book in sample_books:
    print(book)  # Calls __str__

print(repr(sample_books[0]))  # Calls __repr__
print(f"Are the first two books equal? {sample_books[0] == sample_books[1]}")  # False, different ISBNs
    
```
