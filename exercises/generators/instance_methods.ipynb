```python
class Contact(object):
    def __init__(self, first_name, last_name, email):
        self.first_name = first_name
        self.last_name = last_name
        self.email = email
    
    def send_email(self, message):
        print(f"Sending email to {self.email}: {message}")
    
    @property
    def full_name(self):
        return f"{self.first_name} {self.last_name}"
    
    def __str__(self):
        return f"Contact({self.first_name}, {self.last_name}, {self.email})"

alice = Contact("Alice", "Smith", "alice@bob.com")
alice.send_email("Hello Alice!")
print(f"Alice's full name is: {alice.full_name}")
print(alice)
```

```python
class Book(object):
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    
    def taxed_price(self, tax_rate=0.07):
        return self.price *(1.0+tax_rate)

book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 10.99)
print(f"The taxed price of {book1.title} is: ${book1.taxed_price():.2f}")
```
