# Lambda/Anonymous method
1. lambda defines a method but does not assign it to a variable


```python

words = 'this is a bunch of words'.split()
# Sort the words alphabetically
sorted(words)

# sort words by reverse order
sorted(words, key=lambda text: text[::-1])
```

```python
words = 'this is a fantistic bunch of words'.split()
# sort words by length
sorted(words, key=lambda text: sum([1 for char in text if char in 'aeiou']))

sorted(words, key=lambda s: (s.count('p'), s.count('a')))
```
